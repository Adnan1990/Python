### Define Object Oriented Programming Language?

An Object is a thing (Realworld entitiy).Object oriented programming language is a language which deals with objects.Object oriented programming (OOP) aims to implement real world entities like inheritance, hiding, polymorphism etc in programming.

The main aim of OOP is to bind together the data and the functions that operate on them so that no other part of the code can access this data except that function.

Data and its functions are said to be encapsulated into a single entity
Data encapsulation and data hiding are the key terms in the description of object oriented languages

### Differentiate between function and method

#### funtion

A function is a piece of code that is called by name.
It can be passed data to operate on (i.e. the parameters) and can optionally return data (the return value).
All data that is passed to a function is explicitly passed.
Function does not deal with Class and its instance concept.

#### Method 

A method is a piece of code that is called by a name that is associated with an object. In most respects it is identical to a function except for two key differences:

1)A method is implicitly passed the object on which it was called.
2)A method is able to operate on data that is contained within the class (remembering that an object is an instance of a class - the class is the definition, the object is an instance of that data).

### Define the following terms

#### Class

A class is a template for objects.

A Class is a blueprint or template from which objects are created.

A class is a blueprint or prototype that defines the variables and the methods (functions) common to all objects of a certain kind.

A class defines object properties and describes object behavior.

Class is a group of similar objects, Like Human,Fruit,Mobile,Food

Class is declared once.
Class doesn't allocated memory when it is created.


#### Object

Object is an instance of a class, which helps programmers to use variables and methods from inside the class.

Object is a physical entity.

Object is created many times as per requirement.

An Object is a real world entity such as pen, laptop, mobile, bed, keyboard, mouse, chair etc.	

An Object allocates memory when it is created.	


#### Attribute

An Attribute is a named property of a class. It has a type. It describes the range of values that that property may hold.

Attributes are data stores inside class or instance and represent the state or quality of class or instance.

#### Behavior

A class's behavior determines how an instance of that class operates

for example, how it will "react" if asked to do something by another class or object or if its internal state changes. 

Behavior is the only way objects can do anything to themselves or have anything done to them.

Behavior is class member function

```python
class Car:
    def __init__(self,name,color,model,price,mileage):
        self.name = name
        self.color = color
        self.model = model
        self.price = price
        self.mileage = mileage
    
    def update_color(self, color):
        self.color = color
        
    def update_price(self, price):
        self.price = price
        
    def get_details(self):
        print("Name:",self.name,"\nColor:",self.color,"\nModel:",self.model,"\nPrice:",self.price,"\nMileage:",self.mileage)

        
car1 = Car("Ferrari","White",2008,3000000,20)
car2 = Car ("Mehran","Grey",2002,250000,16)
car3 = Car("Civic","Black",2018,205000,18)
car4 = Car("Vitz","Dark Mehroon",2006,400000,16)
car5 = Car("Suzuki","Green",2002,240000,15)

print("\nCar1 Details \n")
car1.get_details()
print("\nCar2 Details \n")
car2.get_details()
print("\nCar3 Details \n")
car3.get_details()
print("\nCar4 Details \n")
car4.get_details()
print("\nCar5 Details \n")
car5.get_details()


car1.update_color("Red")
car1.update_price(260000)

print("\nCar1 Details After updaring color and price \n")
car1.get_details()
```
